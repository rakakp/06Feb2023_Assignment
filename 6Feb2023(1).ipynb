```python
list1 =  [1,2,3,4, [44,55,66, True], False, (34,56,78,89,34), {1,2,3,3,2,1}, {1:34, "key2": [55, 67, 78, 89], 4: (45,22, 61, 34)}, [56, 'data science'], 'Machine Learning']
#list1 = [{1:34, "key2": [55, 67, 78, 89], 4: (45,22, 61, 34)}]

def filter_elements(a):
    final_list = []
    
    for i in a:
        if type(i) == int:
            final_list.append(i)
        elif type(i) == str:
            continue
        elif type(i) == list:
            final_list.append(filter_list(i))
        elif type(i) == bool:
            continue
        elif type(i) == set:
            final_list.append(filter_set(i))
        elif type(i) == tuple:
            final_list.append(filter_tuple(i))
        else:
            final_list.append(filter_dict(i))
    
    print("inside",final_list)
    return final_list
    

def filter_dict(dict4):
    new_list = []
    for k in dict4.keys():
        if type(k) == int:
            new_list.append(k)
        if type(dict4[k]) == int:
            new_list.append(dict4[k])
        if type(dict4[k]) == str:
            continue
        if type(dict4[k]) == list:
            for j in dict4[k]:
                if type(j) == int:
                    new_list.append(j)
        if type(dict4[k]) == tuple:
            for jj in dict4[k]:
                if type(jj) == int:
                    new_list.append(jj)
            
    return new_list
    
def filter_list(lst):
    n_list = []
    for i in lst:
        if type(i) == int:
            n_list.append(i)
    return n_list

def filter_tuple(tup):
    new_list = []
    for i in tup:
        new_list.append(i)
    return new_list

def filter_set(setdata):
    new_list = []
    for i in setdata:
        new_list.append(i)
    return new_list
    
def flat_list(lst):
    lst5 = []
    print(lst)
    for ele in lst:
        if type(ele) == list:
            for item in ele:
                lst5.append(item)
        else:
            lst5.append(ele)
    return lst5
            
    
            
newlst = filter_elements(list1)
print(newlst)
f_list = flat_list(newlst)
from functools import reduce 
print(reduce(lambda x,y: x*y,f_list))
```
